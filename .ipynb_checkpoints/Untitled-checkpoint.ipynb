```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model loaded successfully!")
```

```python
resume = """
Experienced Python developer with 2 years of machine learning experience.
Skilled in pandas, scikit-learn, and building data pipelines.
Worked on classification models and data visualization using matplotlib.
"""

job_description = """
We are looking for a Machine Learning Engineer with strong Python skills.
Candidate must have experience with data pipelines and ML frameworks.
Knowledge of model building and data analysis is required.
"""

print("Resume and job description ready!")
```

```python
resume_embedding = model.encode([resume])
job_embedding = model.encode([job_description])

print("Resume embedding shape:", resume_embedding.shape)
print("Job embedding shape:", job_embedding.shape)
```

```python
score = cosine_similarity(resume_embedding, job_embedding)[0][0]

match_percentage = round(float(score) * 100, 2)

print(f"Match Score: {match_percentage:.2f}%")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Common job posting words that aren't real skills — we'll ignore these
NOISE_WORDS = {
    'looking', 'candidate', 'required', 'must', 'knowledge',
    'experience', 'strong', 'ability', 'including', 'work',
    'skills', 'team', 'position', 'role', 'responsibilities',
    'years', 'good', 'understanding', 'need', 'working'
}

def extract_keywords(text, top_n=10):
    vectorizer = TfidfVectorizer(stop_words='english', max_features=50)
    vectorizer.fit_transform([text])
    
    feature_names = vectorizer.get_feature_names_out()
    scores = vectorizer.idf_
    
    word_scores = sorted(zip(feature_names, scores), key=lambda x: x[1])
    
    # Filter out noise words
    keywords = [
        word for word, score in word_scores 
        if word not in NOISE_WORDS
    ][:top_n]
    
    return keywords

print("keyword extractor ready!")
```

```python
job_keywords = extract_keywords(job_description, top_n=10)
resume_keywords = extract_keywords(resume, top_n=10)

print("Job Description Keywords:")
print(job_keywords)

print("\nResume Keywords:")
print(resume_keywords)
```

```python
def find_keyword_gaps(job_keywords, resume_text):
    # Convert resume to lowercase for fair comparison
    resume_lower = resume_text.lower()
    
    missing_keywords = []
    found_keywords = []
    
    for keyword in job_keywords:
        if keyword.lower() in resume_lower:
            found_keywords.append(keyword)
        else:
            missing_keywords.append(keyword)
    
    return found_keywords, missing_keywords

found, missing = find_keyword_gaps(job_keywords, resume)

print("Keywords Found in Resume:")
print(found)

print("\nKeywords Missing from Resume:")
print(missing)
```

```python
print("=" * 50)
print("       RESUME MATCH REPORT")
print("=" * 50)

print(f"\n Overall Match Score: {match_percentage:.2f}%")

if match_percentage >= 75:
    print("Strong Match!")
elif match_percentage >= 50:
    print("Moderate Match — room to improve")
else:
    print("Weak Match — significant gaps found")

print(f"\n Skills Found ({len(found)}):")
for skill in found:
    print(f"   • {skill}")

print(f"\n Missing Keywords ({len(missing)}):")
for skill in missing:
    print(f"   • {skill}")

print(f"\n Tip: Add these keywords to your resume to improve your score!")
print("=" * 50)
```

```python
import fitz

def read_pdf(file_path):
    
    pdf = fitz.open(file_path)
    
    full_text = ""

    for page in pdf:
        full_text += page.get_text()
    
    pdf.close()
    
    return full_text.strip()

print("PDF reader ready!")
```

```python
resume_text = read_pdf('Resume.pdf')

print(resume_text[:500])
```

```python
resume = read_pdf('Resume.pdf')

job_description = """
We are looking for a Machine Learning Engineer with strong Python skills.
Candidate must have experience with data pipelines and ML frameworks.
Knowledge of model building and data analysis is required.
"""

resume_embedding = model.encode([resume])
job_embedding = model.encode([job_description])

score = cosine_similarity(resume_embedding, job_embedding)[0][0]
match_percentage = round(float(score) * 100, 2)

job_keywords = extract_keywords(job_description, top_n=10)
found, missing = find_keyword_gaps(job_keywords, resume)

print("=" * 50)
print("       RESUME MATCH REPORT")
print("=" * 50)

print(f"\n📊 Overall Match Score: {match_percentage:.2f}%")

if match_percentage >= 75:
    print("Strong Match!")
elif match_percentage >= 50:
    print("Moderate Match — room to improve")
else:
    print("Weak Match — significant gaps found")

print(f"\n Skills Found ({len(found)}):")
for skill in found:
    print(f"   • {skill}")

print(f"\n Missing Keywords ({len(missing)}):")
for skill in missing:
    print(f"   • {skill}")

print(f"\n Tip: Add these keywords to your resume to improve your score!")
print("=" * 50)
```
